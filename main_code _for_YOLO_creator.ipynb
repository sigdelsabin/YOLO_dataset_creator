```python
"""
Batch YOLO Dataset Creator
=============================================
Process all TIF files with the same shapefile using YoloDatasetFromShp class.
"""

import os
from pathlib import Path
from yolo_dataset_creator import YoloDatasetFromShp

# ============================================================
# CONFIGURATION
# ============================================================
TIF_DIRECTORY = r"C:\IFA\2025_Datasets\composites\thermal_3band"
SHAPEFILE_PATH = r"C:\IFA\2025_Datasets\8_12_4\labels_shp\labels_12pm_rgb.shp"
OUTPUT_BASE_DIR = r"C:\IFA\2025_Datasets\YOLO_Datasets_thermal_stack"  # Change this to your desired output location

# Dataset parameters
CROP_SIZE = 640
TRAIN_RATIO = 0.8
VALID_RATIO = 0.1
TEST_RATIO = 0.1
CLASS_NAMES = ["ifa_mound"]
RANDOM_SEED = 42

# ============================================================
# MAIN
# ============================================================

if __name__ == "__main__":
    
    # Create output base directory
    os.makedirs(OUTPUT_BASE_DIR, exist_ok=True)
    
    # Find all TIF files
    tif_files = list(Path(TIF_DIRECTORY).glob("*.tif")) + \
                list(Path(TIF_DIRECTORY).glob("*.TIF"))
    
    print(f"\n{'='*70}")
    print(f"Found {len(tif_files)} TIF file(s) to process")
    print(f"{'='*70}\n")
    
    # Process each TIF file
    for idx, tif_file in enumerate(tif_files, 1):
        print(f"[{idx}/{len(tif_files)}] Processing: {tif_file.name}\n")
        
        output_dir = os.path.join(OUTPUT_BASE_DIR, f"Dataset_{tif_file.stem}")
        
        try:
            creator = YoloDatasetFromShp(
                input_tif=str(tif_file),
                input_shp=SHAPEFILE_PATH,
                output_dir=output_dir,
                crop_size=CROP_SIZE,
                train_ratio=TRAIN_RATIO,
                valid_ratio=VALID_RATIO,
                test_ratio=TEST_RATIO,
                class_names=CLASS_NAMES,
                random_seed=RANDOM_SEED
            )
            
            creator.create_dataset()
            
        except Exception as e:
            print(f"\n Error: {str(e)}\n")
        
        print(f"{'='*70}\n")
    
    print(f"âœ… All datasets saved to: {OUTPUT_BASE_DIR}\n")
```
